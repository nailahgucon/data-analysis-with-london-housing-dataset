```python
import pandas as pd
```

```python
housing_df = pd.read_csv("london-housing-data.csv")
```

```python
housing_df
```

## 1. Convert the data type of 'Date' column to date-time format.

```python
housing_df.dtypes
```

```python
housing_df.date = pd.to_datetime(housing_df.date)
```

```python
housing_df.dtypes
```

## 2. Add a new column 'year' in the dataframe which contains years only.

```python
housing_df['year'] = housing_df.date.dt.year
```

```python
housing_df
```

#### Extra:

#### Rearrange the sequence of the columns to start with the 'year' column

```python
housing_df = housing_df.reindex(columns=['year','date','area','average_price','code','houses_sold','no_of_crimes'])
```

```python
housing_df
```

#### Alternatively...

```python
# reapply year to the end
housing_df.drop(['year'], axis=1, inplace = True)
housing_df['year'] = housing_df.date.dt.year
housing_df
```

```python
cols = housing_df.columns.tolist()
cols
```

```python
# move the last element to the first position
cols = cols[-1:] + cols[:-1]
cols
```

```python
housing_df = housing_df[cols]
housing_df
```

#### Alternatively...

```python
# reapply year to the end
housing_df.drop(['year'], axis=1, inplace = True)
housing_df['year'] = housing_df.date.dt.year
housing_df
```

```python
housing_df.drop('year', axis=1, inplace=True) 
housing_df.head()
```

```python
housing_df.rename(columns={'year_temp': 'year'}, inplace=True)
housing_df.head()
```

## 3. To add a month as a new column, between date and area columns.

```python
housing_df.head()
```

```python
housing_df.insert(1, 'month', housing_df.date.dt.month)
```

```python
housing_df
```

## 4. Remove the columns year and month from the dataframe.

```python
housing_df.drop(['year','month'], axis=1, inplace = True)
```

```python
housing_df
```

## 5. Show all the records where 'No of crimes' are zero and how many such records are there?

```python
housing_df[housing_df.no_of_crimes == 0]
```

```python
len(housing_df[housing_df.no_of_crimes == 0])
```

#### Extra:

#### What can you tell about how the no_of_crimes affect other factors such as houses_sold and average_price?

```python
housing_df.plot(x="no_of_crimes", y="houses_sold", kind="hist")
```

#### More houses were sold where the least amount of crime occured.

```python
housing_df.plot(x="no_of_crimes", y="average_price", kind="hist")
```

#### The lesser the crime occuring, the more expensive the house

## 6. What is the maximum and minimum 'average_price' per year in England?

```python
housing_df['year'] = housing_df.date.dt.year
```

```python
housing_df
```

```python
england_housing_df = housing_df[housing_df.area == 'england']
```

```python
england_housing_df
```

```python
england_housing_df.groupby('year').average_price.max()
```

```python
england_housing_df.groupby('year').average_price.mean()
```

## 7. What is the maximum and minimum no of crimes recorded area?

```python
housing_df.groupby('area').no_of_crimes.max()
```

```python
housing_df.groupby('area').no_of_crimes.min().sort_values(ascending = True)
```

## 8. Show the total count of records of each area, where average price is less than 100000.

```python
housing_df[housing_df.average_price < 100000].area.value_counts()
```
